## 1. Introducción

La predicción de fraude en las empresas en un problema conocido y bastante estudiado debido al interés que supone sobre minizar los riesgos que adoptan las empresas con sus clientes. El problema que se presenta en el proyecto consta de detección de fraude de cliente por parte de banco. 

La forma de abordar el problema consistirá de emplear modelos de clasificación (donde 1 correspondería a cliente clasificado como fraudulento y 0 como cliente no fraudulento) en los diferentes segmentos etarios de la empresa. Los cuales son 4:

* Joven (Menor a 18 años)
* Adulto-Joven (18 - 26 años)
* Adulto (27- 59 años)
* Persona Mayor (60 años o mas).

Donde los datos de entrada del modelo corresponden a datos asociados al cliente donde algunas columnas tienen datos nulos y sus representaciones corresponden a valores negativos, -1 o valores NA. 

En cuanto a la hipótesis del problema, se usará un enfoque donde el aprendizaje de máquinas tiene todo el poder predictivo, o sea en otras palabras, se le entregarán todas las variables que cuenta el dataset y el modelo de ML será el que le de el peso a las variables que son más importantes para predecir este problema. En cuanto a los resultados esperados, no se espera que sean muy ambiciosos debido a que es un típico problema donde hay mucho desbalance de clases y los modelos tenderán a predecir la clase mayoritaria.

## 6. Anomalías

### 6.1 Implementar detector de anomalías sobre dataframe de muestreo

La forma en que se detectaran anomalías por rango etario serán usando modelos de clasificación. Se probarán base dummy, decision tree y knn. Los modelos que presenten un mejor recall y precisión serán los escogidos, dada la naturaleza del problema. Esto debido a que el recall es la proporción de la clase positiva que efectivamente fueron detectedas con éxito, debido al desbalance de clases existente, sería la mejor métrica a considerar.

```python
from sklearn.dummy import DummyClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.svm import OneClassSVM
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score
from sklearn.model_selection import train_test_split
import numpy as np

def run_classifier(clf, X, y):
        metrics = {'f1-score': [], 'precision': [], 'recall': []}
    
        #dividiendo los datos de entrenamiento y validación
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30,  random_state=101)

        clf.fit(X_train,y_train) #fiteamos el modelo
        predictions = clf.predict(X_test) #Generamos las predicciones
       
        
        metrics['y_pred'] = predictions
        metrics['f1-score'].append(f1_score(y_test, predictions)) 
        metrics['recall'].append(recall_score(y_test, predictions))
        metrics['precision'].append(precision_score(y_test, predictions))
        matriz= confusion_matrix(y_test, predictions)


        return metrics, matriz
```

#### Menor

```python
X=menor[['comp1','comp2']]
y=menor['fraud_bool']

c0 = ("Base Dummy", DummyClassifier(strategy='stratified', random_state=10))
c1 = ("Decision Tree", DecisionTreeClassifier(max_depth=5))
c2 = ("KNN", KNeighborsClassifier(n_neighbors=10))

classifiers = [c0, c1, c2]

results = {}
for name, clf in classifiers:
    metrics, matriz = run_classifier(clf, X, y)   # hay que implementarla en el bloque anterior.
    results[name] = metrics
    print("----------------")
    print("Resultados para clasificador: ", name) 
    print(matriz)
    print("Precision promedio:", np.array(metrics['precision']).mean())
    print("Recall promedio:", np.array(metrics['recall']).mean())
    print("F1-score promedio:", np.array(metrics['f1-score']).mean())
    print("----------------\n\n")  
```
